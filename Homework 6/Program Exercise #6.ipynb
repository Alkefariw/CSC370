```python
# /* Global declarations */
# /* Variables */
charClass = int()
# //this is to detect floating point literals
prevClass = int()
lexeme = list()
lexeme =  [''] * 100
nextChar = ''
nextCharPlus = ''
lexLen = int()
token = int()
nextToken = int()
# //the goal is to look up and compare to see if the token that comes after will be a '*'
prevToken = 0
# Character classes
LETTER = 0
DIGIT = 1
COMMENT = 2
UNKNOWN = 99
INT_LIT = 10
IDENT = 11
ASSIGN_OP = 20
ADD_OP = 21
SUB_OP = 22
MULT_OP = 23
DIV_OP = 24
LEFT_PAREN = 25
RIGHT_PAREN = 26
EOF = ''
in_fp = open("front.in", "r")

def addChar():
    global lexLen
    global nextChar
    global lexeme
    if (lexLen <= 98):
        lexeme[lexLen] = nextChar
        lexLen += 1

        lexeme[lexLen] = 0
    else:
        print("Error - lexeme is too long \n")

def getChar():
    global charClass, nextChar
    nextChar = in_fp.read(1)
    if nextChar != EOF:
        if nextChar.isalpha():
            charClass = LETTER
        elif nextChar.isdigit():
            charClass = DIGIT
        else:
            charClass = UNKNOWN
    else:
        charClass = EOF

def lookup(ch):
    global prevToken
    global nextToken
    prevToken = nextToken
    if ch == '(':
        addChar()
        nextToken = LEFT_PAREN
    elif ch == ')':
        addChar()
        nextToken = RIGHT_PAREN
    elif ch == '+':
        addChar()
        nextToken = ADD_OP
    elif ch == '-':
        addChar()
        nextToken = SUB_OP
    elif ch == '*':
        addChar()
        nextToken = MULT_OP
    elif ch == '/':
        addChar()
        nextToken = DIV_OP
    else:
        addChar()
        nextToken = EOF
    return nextToken
def getNonBlank():
    global nextChar
    while nextChar.isspace():
        getChar()


def lex():
    global prevClass, nextToken, charClass, lexeme, prevToken, lexLen

    lexLen = 0
    getNonBlank()
    if charClass == LETTER:
        addChar()
        getChar()
        while (charClass == LETTER or charClass == DIGIT):
            addChar()
            prevClass = charClass
            getChar()
            nextToken = IDENT
    elif charClass == DIGIT:
        addChar()
        getChar()
        while (charClass == DIGIT):
            addChar()
            getChar()
            nextToken = INT_LIT
    elif charClass == UNKNOWN:
        lookup(nextChar)
        getChar()
        if (prevToken == 24 and nextToken == 23):
            addChar()
            getChar()
        elif (prevToken == 23 and nextToken == 24): 
            addChar()
            getChar()
    elif charClass == EOF:
        nextToken = EOF
        lexeme[0] = 'E'
        lexeme[1] = 'O'
        lexeme[2] = 'F'
        lexeme[3] = 0
    str_lex = ''
    for i in range(len(lexeme)):
        if lexeme[i]!= 0:
            str_lex += lexeme[i]
        elif lexeme[i] == 0:
            break
    print("Next token is:", -1 if nextToken == EOF else nextToken, "Next lexeme is:", str_lex)
    return nextToken
def main():
    global nextToken
    in_fp = open("front.in", "rb")
    if not in_fp:
        print("ERROR - cannot open front.in \n")
    else: 
        getChar()
        while (nextToken != EOF):
            lex()
main()
```
