```python
## Program with Hash function:-
import random

def generateName():
    character_array = list()
    for i in range(65, 91):
        character_array.append(chr(i))
    name = ''
    for i in range(3):
        index = random.randrange(0, len(character_array) - 1)
        name += character_array[index]
    return name

def generateAge():
    return random.randrange(1, 99)

hashmap = {}

name1 = generateName()
print('First name:', name1)

name2 = generateName()
print('Second name:', name2)
if name2 != name1:
    pass
else:
    while(name2 == name1):
        print('Duplicate second name:', name2)
        name2 = generateName()
        print('New second name:', name2)

name3 = generateName()
print('Third name:', name3)
if name3 != name1 and name3 != name2:
    pass
else:
    while(name3 == name1 or name3 == name2):
        print('Duplicate third name:', name3)
        name3 = generateName()
        print('New third name:', name3)

age1 = generateAge()
hashmap[name1] = age1
age2 = generateAge()
hashmap[name2] = age2
age3 = generateAge()
hashmap[name3] = age3

for keys, values in hashmap.items():
    print(keys, ",", values)
```

```python
## Program without Hash function:-
import random

def generateName():
    character_array = list()
    for i in range(65, 91):
        character_array.append(chr(i))
    name = ''
    for i in range(3):
        index = random.randrange(0, len(character_array) - 1)
        name += character_array[index]
    return name

def generateAge():
    return random.randrange(1, 99)

name = list()
name1 = generateName()
print('First name:', name1)
name.append(name1)

name2 = generateName()
print('Second name:', name2)
if name2 != name1:
    name.append(name2)
else:
    while(name2 != name1):
        print('Duplicate second name:', name2)
        name2 = generateName()
        print('New second name:', name2)
    name.append(name2)

name3 = generateName()
print('Third name:', name3)
if name3 != name1 and name3 != name2:
    name.append(name3)
else:
    while(name3 != name1 and name3 != name2):
        print('Duplicate third name:', name3)
        name3 = generateName()
        print('New third name:', name3)
    name.append(name3)


age = list()
age1 = generateAge()
age.append(age1)
age2 = generateAge()
age.append(age2)
age3 = generateAge()
age.append(age3)

for i in range(3):
    print(name[i],'with age:', age[i])
```
